## Bar Charts

```python
google = data.DataReader(name = "GOOG", data_source = "google", start = "2004-01-01", end = "2017-01-01")
google.head()
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "barh")
```

## Pie Charts

```python
apple = data.DataReader(name = "AAPL", data_source = "google", start = "2012-01-01", end = "2016-12-31")
apple.head()
```

```python
apple["Close"].mean()
```

```python
def rank(stock_price):
    if stock_price >= 93.36:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("ggplot")
apple["Close"].apply(rank).value_counts().plot(kind = "pie", legend = True)
```

## Histograms

```python
google = data.DataReader(name = "GOOG", data_source = "google", start = "2004-01-01", end = "2017-01-01")
google.head(3)
```

```python
def buckets(price):
    return int(price / 100) * 100
```

```python
google["High"].apply(buckets).value_counts().sort_index()
```

```python
google["High"].apply(buckets).nunique()
```

```python
google["High"].apply(buckets).plot(kind = "hist", bins = 9)
```
